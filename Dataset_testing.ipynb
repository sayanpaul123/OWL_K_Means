# Brain

```python
os.chdir("../Data_microarray")
```

```python
dataset = np.loadtxt('brain.x.txt')
label = np.loadtxt('brain.y.txt')
num_clust = np.unique(label).shape[0]  
beta = 4
```

```python
print(dataset.shape)
```

```python
nmi_list = []
for _ in range(10):
    computed, weight = owlk(cp.array(dataset.T),num_clust = num_clust,beta = beta,nu=0,sigma=1) ## Calling the function
    nmi_list.append(nmi(computed,label))
    print(nmi(computed,label))
```

```python
from statistics import mean,variance,stdev
print("Mean NMI:",mean(nmi_list))
print("std. NMI:",stdev(nmi_list))
```

```python
beta = 4
nmi_list = []
ari_list = []
for _ in range(10):
    computed, weight = owlk(cp.array(dataset.T),num_clust = num_clust,beta = beta,nu=0,sigma=1,rev = True) ## Calling the function
    nmi_list.append(nmi(computed,label))
    ari_list.append(ari(computed,label))
    print(nmi(computed,label))
```

```python
from statistics import mean,variance,stdev
print("Mean NMI:",mean(nmi_list))
print("std. NMI:",stdev(nmi_list))
print("Mean ARI:",mean(ari_list))
print("std. ARI:",stdev(ari_list))
```
